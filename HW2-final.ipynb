```python
#Q1 資料中哪間校系的延畢人數最多

max_delayed = df['延修生總計'].max()

delayed_dept = df.loc[df['延修生總計'] == max_delayed, '校系'].iloc[0]

print(f"延畢人數最多的校系為：{delayed_dept}，人數為：{max_delayed}")
```

```python
#Q2 哪間校系的學生人數從大四到大一人數平均增加率最高及最低

# 計算增加率
df['增加率(大四到大三)'] = (df['三年級總計'] - df['四年級總計']) / df['四年級總計'] * 100
df['增加率(大三到大二)'] = (df['二年級總計'] - df['三年級總計']) / df['三年級總計'] * 100
df['增加率(大二到大一)'] = (df['一年級總計'] - df['二年級總計']) / df['二年級總計'] * 100

# 計算平均增加率
df['平均增加率'] = (df['增加率(大四到大三)'] + df['增加率(大三到大二)'] + df['增加率(大二到大一)']) / 3

# 找出平均增加率最高的校系
max_avg_growth = df['平均增加率'].max()
max_avg_growth_dept = df.loc[df['平均增加率'] == max_avg_growth, '校系'].iloc[0]

print(f"從大四到大一人數平均增加率最高的校系為：{max_avg_growth_dept}，平均增加率為：{max_avg_growth}%")

# 找出平均增加率最低的校系
min_avg_growth = df['平均增加率'].min()
min_avg_growth_dept = df.loc[df['平均增加率'] == min_avg_growth, '校系'].iloc[0]

print(f"從大四到大一人數平均增加率最低的校系為：{min_avg_growth_dept}，平均增加率為：{min_avg_growth}%")
```

```python
#Q3 哪間校系的各年級學生人數最穩定?

df = df_student.loc[filt, ['學校名稱','科系名稱','一年級總計','二年級總計','三年級總計','四年級總計','五年級總計','六年級總計','七年級總計','延修生總計']]

df_sorted = df.sort_values(by='一年級總計', ascending=False)

top_30 = df_sorted.head(30)

df = top_30

def sum_ab(row):
    return row['學校名稱'] + row['科系名稱']

df['校系'] = df.apply(sum_ab,axis=1)

import numpy as np

# 計算各年級學生人數的變異程度
df['變異程度'] = np.var(df[['一年級總計', '二年級總計', '三年級總計', '四年級總計']], axis=1)

# 找到變異程度最小的校系
most_stable_department = df.loc[df['變異程度'].idxmin(), '校系']

print(f"各年級學生人數最穩定的校系為：{most_stable_department}")
```
