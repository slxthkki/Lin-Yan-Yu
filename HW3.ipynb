```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# 目標網頁的URL
url = 'https://www.ptt.cc/bbs/Food/M.1559281587.A.E01.html'
res = requests.get(url)
html = res.text
soup = BeautifulSoup(html, "html.parser")
```

```python
# 獲取文章標題
title = soup.title.string
print(title)
```

```python
print(soup.prettify())
```

```python
# 使用正則化表達式擷取數字後的咖啡店名稱

text = soup.prettify()

pattern = r"(\d+\.\s)(.*?)(?=\n)"

matches = re.findall(pattern, text, re.DOTALL)

for match in matches:
    number = match[0]
    content = match[1]
    print(number, content)
```

```python
# 使用正則化表達式擷取咖啡店名稱、地址、營業時間

text = soup.prettify()

pattern = r"(\d+\.\s)(.*?)(?=\n)|地址：(.*?)(?=\n)|營業時間：(.*?)(?=\n)"

matches = re.findall(pattern, text, re.DOTALL)

result = ""
for match in matches:
    number = match[0]
    content = match[1]
    address = match[2]
    opening_hour = match[3]
    result += f"{number} {content}\n"
    result += f"地址: {address}\n"
    result += f"營業時間：{opening_hour}\n"
    
print(result)
```

```python
import json

text = soup.prettify()

pattern = r"(\d+\.\s)(.*?)(?=\n)|地址：(.*?)(?=\n)|營業時間：(.*?)(?=\n)"

matches = re.findall(pattern, text, re.DOTALL)

data = []
for match in matches:
    number = match[0]
    content = match[1]
    address = match[2]
    opening_hour = match[3]
    
    shop_data = {
        "number": number.strip(),
        "content": content.strip(),
        "address": address.strip(),
        "opening_hour": opening_hour.strip()
    }
    
    data.append(shop_data)
    
# 存成JSON文件
with open("coffee_shops.json", "w", encoding="utf-8") as json_file:
    json.dump(data, json_file, ensure_ascii=False, indent=4)
```
